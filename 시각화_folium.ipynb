```python
import folium
```

```python
# 지도 시각화 ; 지도 생성 -> 정보 추가

# 지도 생성 ; folium.Map(location = [위도, 경도], zoom_start=확대정도)
# 서울역
m = folium.Map(location=[37.5536067,126.9674308], zoom_start=15)
m

# 마커 추가 ; folium.Marker([위도, 경도].add_to(m;생성한 지도))
folium.Marker([37.5536067,126.9674308]).add_to(m)

# 정보 추가
folium.Marker([37.5536067,126.9674308],
             tooltip='mouseover', # 커서 설명
             popup='click' # 커서 클릭하면 설명 나옴
             ).add_to(m)
m

# 정보 추가 써클 마커
folium.CircleMarker([37.5536067,126.9674308],
                   radius=20, # 원의 크기
                   tooltip='mouseover_circlemarker').add_to(m)
m
```

```python
# 미니맵 추가

from folium.plugins import MiniMap
m = folium.Map(location = [37.5536067,126.9674308], zoom_start=12)

# 미니맵 추가
minimap = MiniMap()
minimap.add_to(m)
m
```

```python
# 서울 대피소 현황 지도 만들기 실습

import pandas as pd
file = './data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'

raw = pd.read_csv(file, encoding='cp949')
raw.head()
# raw.info()
```

```python
# raw.iloc[0,-2:]
# raw.loc[0,['위도', '경도']]

lat = raw.loc[0,'위도']
long = raw.loc[0,'경도']

print(lat, long)
```

```python
lat = raw.loc[0,'위도']
long = raw.loc[0,'경도']
name = raw.loc[0, '대피소명칭']

print(lat, long, name)
```

```python
for i in raw.index:
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    print(lat, long, name)
```

```python
# 지도 생성 후 마커 추가
# 중심 : 서울역[37.5536067,126.9674308]

m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
m

# 대피소 마커 추가
for i in raw.index:
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long],
                 tooltip=name).add_to(m)
    
m
```

```python
# 지도 저장
m.save('./data/ShelterMap_Seoul.html')
```

```python
# MarkerCluster
from folium.plugins import MarkerCluster
```

```python
# 클러스터 마커 ; 마커 그룹화

# 지도 생성 후 마커 추가
# 중심 : 서울역[37.5536067,126.9674308]

m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)

# 마커 클러스터에 추가
for i in raw.index:
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long],
                 tooltip=name).add_to(marker_cluster)
    
m
```

```python
# 미니맵 추가
from folium.plugins import MiniMap

m = folium.Map(location=[37.5536067,126.9674308],
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)

minimap = MiniMap()
minimap.add_to(m)

# 마커 클러스터에 추가
for i in raw.index:
    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long],
                 tooltip=name).add_to(marker_cluster)
    
m
```

```python
m.save('./data/ShelterMap_Seoul_Cluster.html')
# End 221202
```
